```python
import numpy as np

def ulam_spiral(size):
    grid = [[0] * size for _ in range(size)]
    directions = [(0, 1), (-1, 0), (0, -1), (1, 0)]
    direction = 0  # start moving to the right
    x, y = size // 2, size // 2  # start from the center
    num = 1
    step_size = 1
    step_count = 0
    while num <= size * size:
        grid[x][y] = num
        num += 1
        x += directions[direction][0]
        y += directions[direction][1]
        step_count += 1
        if step_count == step_size:
            step_count = 0
            direction = (direction + 1) % 4
            if direction == 0 or direction == 2:
                step_size += 1
    return grid

def print_ulam_spiral(grid):
    horizontal_line = "+-----" * len(grid[0]) + "+"
    for row in grid:
        print(horizontal_line)
        for num in row:
            print(f"|{num:^5}", end="")
        print("|")
    print(horizontal_line)

size = 11
grid = ulam_spiral(size)
print_ulam_spiral(grid)
```
