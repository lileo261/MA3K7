```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

A = np.array([1,0.5,0.75])
for i in np.arange(3,25):
    a = 0.5*A[i-1] + 0.5*A[i-2]
    A = np.append(A,a)

x = np.arange(1, 1 + len(A))

print(A)

plt.plot(x, A)
plt.title('Probability of landing on a given step')
plt.ylim(0)
plt.grid('on')
plt.xlabel('Step')
plt.ylabel('Probability')
plt.show()
```

```python
B = np.array([1,0.5,0.75])
for i in np.arange(3,25):
    b = 1 - 0.5*A[i-1]
    B = np.append(B,b)

x = np.arange(1, 1 + len(B))

print(B)

plt.plot(x, B)
plt.title('Probability of landing on a given step')
plt.ylim(0)
plt.grid('on')
plt.xlabel('Step')
plt.ylabel('Probability')
plt.show()
```
